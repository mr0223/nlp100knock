```python
# 60. 単語ベクトルの読み込みと表示
import gensim
model = gensim.models.KeyedVectors.load_word2vec_format('data/ch07/GoogleNews-vectors-negative300.bin', binary=True)

print('vector of "United States"')
print(model['United_States'])
```

```python
# 61. 単語の類似度
c_similarity = model.similarity('United_States', 'U.S.')

print(f'Cosine similarity between United_States and U.S.: {c_similarity}')
```

```python
# 62. 類似度の高い単語10件
similar_words = model.most_similar(positive='United_States', topn=10)

print('10 most similar words to United_States and their similarities')
for similar_word in similar_words:
    print(similar_word)
```

```python
# 63. 加法構成性によるアナロジー
analogous_words = model.most_similar(positive=['Spain', 'Athens'], negative=['Madrid'], topn=10)

print('analogous 10 words')
for analogous_word in analogous_words:
    print(analogous_word)
```

```python
# 64. アナロジーデータでの実験
with open('data/ch07/questions-words.txt') as f:
    lines = f.readlines()

for line in lines:
    if line[0]==':':
        with open('data/ch07/answer-words.txt', 'a', encoding='utf-8') as f:
            f.write(line)
        continue
    
    # 2列目の単語 - 1列目の単語 + 3列目の単語の計算
    tmp_words = line.split()
    analogous_word, similarity = model.most_similar(positive=[tmp_words[1], tmp_words[2]], negative=tmp_words[0], topn=1)[0]
    
    with open('data/ch07/answer-words.txt', 'a', encoding='utf-8') as f:
        f.write(f'{" ".join(tmp_words)} {analogous_word} {similarity}\n')
```

```python
# 65. アナロジータスクでの正解率
with open('data/ch07/answer-words.txt') as f:
    lines = f.readlines()

counts   = [0, 0]  # 意味的アナロジー, 文法的アナロジーのカウント
corrects = [0, 0]  # 意味的アナロジー, 文法的アナロジーの正解数

flag = 0           # 意味的アナロジー, 文法的アナロジーを区別

for line in lines:
    if 'gram' in line:   # 'gram'の入った行で意味と文法が分かれる
        flag = 1
    if line[0] == ':':   # '：'で始まる行はカウントしない
        continue
    
    tmp_words = line.split()   # 単語に分割
    
    if tmp_words[3] == tmp_words[4]:
        corrects[flag] += 1
    counts[flag] += 1

print(f'semantic analogy accuracy : {corrects[0] / counts[0]}({corrects[0]}/{counts[0]})')
print(f'syntactic analogy accuracy: {corrects[1] / counts[1]}({corrects[1]}/{counts[1]})')
```

```python
# 66. WordSimilarity-353での評価
import pandas as pd
from scipy.stats import spearmanr

# WordSimilarity-353の読み込み
human_df = pd.read_csv('data/ch07/combined.csv')

# 単語ベクトルによる類似度の追加
vector_column = []
for row in human_df.itertuples():
    vector_column.append(model.similarity(row._1, row._2))
human_df['vector'] = vector_column

# 類似度ランキングの追加
human_df['human_rank'] = human_df.rank(ascending=False)['Human (mean)']
human_df['vector_rank'] = human_df.rank(ascending=False)['vector']

# スピアマン相関係数の計算
corr, pvalue = spearmanr(human_df['human_rank'], human_df['vector_rank'])

print(f'Spearman\'s rank correlation coefficient: {corr}')
```

```python
# 67. k-meansクラスタリング
import numpy as np
from sklearn.cluster import KMeans

# 国名に関する単語ベクトルを抽出
# 国名のリスト:data/ch07/countries.txt(https://gist.github.com/kalinchernev/486393efcca01623b18dを利用し, 学習済みの語彙に対応, 195か国)
def extract_country_vecs():
    '''
    input : なし
    output: country_names(list(195,)), vecs(nparray(195, 300))
    '''
    # 国名の抽出
    with open('data/ch07/countries.txt') as f:
        country_data = f.read()
    country_names = country_data.split()
    
    # 国名の単語ベクトルを抽出
    country_vecs = []
    for country_name in country_names:
        country_vecs.append(model[country_name])
    
    return country_names, np.array(country_vecs)

country_names, country_vecs = extract_country_vecs()

# k-means(k=5)
pred = KMeans(n_clusters=5).fit_predict(country_vecs)

for i in range(5):
    cluster_members = [country_name for num, country_name in enumerate(country_names) if pred[num]==i]
    print(f'cluster{i}: {"  ".join(cluster_members)}\n')
```

```python
# 68. Ward法によるクラスタリング
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# 国名に関する単語ベクトルを抽出: def in 67
country_names, country_vecs = extract_country_vecs()

# Ward法による階層型クラスタリング
result = linkage(country_vecs, method='ward')

# 結果をデンドログラムとして可視化
plt.figure(figsize=(20,10))
dendrogram(result, labels=country_names)
plt.show()
```

```python
# 69. t-SNEによる可視化
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# 国名に関する単語ベクトルを抽出: def in 67
country_names, country_vecs = extract_country_vecs()

# ベクトル空間上の単語ベクトルをt-SNEで可視化
country_vecs_embedded = TSNE().fit_transform(country_vecs)

plt.figure(figsize=(20,10))
plt.title('word vectors for country names')
plt.scatter(country_vecs_embedded.T[0], country_vecs_embedded.T[1])
for i, label in enumerate(country_names):
    plt.annotate(label, country_vecs_embedded[i])
plt.show()
```
